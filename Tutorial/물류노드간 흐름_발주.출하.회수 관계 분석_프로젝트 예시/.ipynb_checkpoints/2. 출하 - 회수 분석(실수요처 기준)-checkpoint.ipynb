```python
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams["font.family"] = 'Malgun Gothic'
```

```python
df1 = pd.read_csv('./2-1. 회수(실수요처 5개).csv', engine='python', encoding='utf-8')
df2 = pd.read_csv('./2-2. 출하(실수요처 5개).csv', engine='python', encoding='utf-8')
```

```python
df1['출발지'].unique()
```

```python
df1['수량'].sum()
```

```python
df2['도착지'].unique()
```

```python
df2['수량'].sum()
```

### 실수요처 관계 분석 코드

- 5115152: 상위 1위
- 5083226: 상위 100위
- 5113634: 상위 500위
- 5152378: 상위 1000위
- 5214623: 상위 5000위

```python
df1['날짜'] = pd.to_datetime(df1['날짜'], format="%Y%m%d")
df2['날짜'] = pd.to_datetime(df2['날짜'], format="%Y%m%d")
```

# 1. 실수요처 상위 1위 - 5115152

```python
df11 = df1[df1['출발지'] == 5115152]
df12 = df2[df2['도착지'] == 5115152]
```

```python
df11a = df11.groupby('날짜')[['수량']].sum()
df12a = df12.groupby('날짜')[['수량']].sum()

df11a = df11a.reindex(pd.date_range(pd.to_datetime('2010-01-01'), pd.to_datetime('2018-06-30'), freq='d')).fillna(0)
df12a = df12a.reindex(pd.date_range(pd.to_datetime('2010-01-01'), pd.to_datetime('2018-06-30'), freq='d')).fillna(0)

df_first = pd.concat([df11a, df12a], axis=1)
df_first.columns = ['회수', '출하']
```

```python
df_first.resample('m').sum().tail()
```

## 1-1. 연 기준

```python
df_first.resample('1Y').sum().plot(figsize=(10,10))
```

### 흠

## 1-2. 월 기준

```python
df_first['2015':'2016'].resample('1M').sum().plot(figsize=(10,10))
```

### 같이 커지고 작아지는 것 같은데 흠

## 3. 주 기준

```python
df_first['2015-02':'2015-06'].resample('1W').sum().plot(figsize=(10,10))
```

### 같이 커지고 작아지는 것 같은데 흠
